```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv('vgsales.csv')
```

```python
plt.plot(df['Year'])
plt.show()
```

```python
plt.scatter(x = df['Year'], y = df['NA_Sales'])
plt.show()
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(df['JP_Sales'], color='red')
plt.ylabel('Sales')
plt.title('Japan Sales')

plt.subplot(1, 2, 2)
plt.plot(df['NA_Sales'], color='purple')
plt.ylabel('Sales')
plt.title('N America Sales')
plt.show()
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.scatter(x = df['Year'], y = df['JP_Sales'], color='red')
plt.ylabel('Sales')
plt.title('Japan Sales')

plt.subplot(1, 2, 2)
plt.scatter(x = df['Year'], y = df['NA_Sales'], color='purple')
plt.ylabel('Sales')
plt.title('N America Sales')
plt.show()
```
